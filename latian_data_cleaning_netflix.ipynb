```python
import pandas as pd
```

```python
df = pd.read_csv(r"D:\Dhika Ramadhan Saputra\Data Analist\Kebutuhan\netflix_titles.csv")
df
```

```python
#Identify how many null values every column
df.isnull().sum().sort_values(ascending=False)
```

```python
#identify how many percentage null values for each column
for column in df.columns:
    percentage = str(round(df[column].isnull().mean()*100, 2))
    print(column, percentage)
```

```python
#drop director row that contains null values
director_null_rows = df[df['director'].isnull()].index
df.drop(director_null_rows, inplace = True)
df
```

```python
df.isnull().sum()
```

```python
#df[df['rating'].isnull()]
mode = ''.join(df['rating'].mode())
df['rating'].fillna(mode, inplace=True)
```

```python
df.isnull().sum()
```

```python
df[df['cast'].isnull()]
```

```python
df_movie = df[df['type'] == 'Movie']
df_movie
```

```python
df_movie['minute'] = df_movie['duration'].str.split(expand=True)[0].astype(int)
df_movie
```

```python
df_movie['released_year'] = df_movie['date_added'].str.extract('(\d{4})').astype(int)
df_movie
```

```python
df_movie.drop(columns=['date_added', 'duration'], inplace=True)
df_movie
```

```python
df_movie.drop(df_movie.columns[6], axis=1, inplace=True)
df_movie
```
